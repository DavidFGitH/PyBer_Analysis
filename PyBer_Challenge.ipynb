### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
#David Comment: For this df and next few, I use the groupby function by "type" which combines all cities together and count by what I need
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
#David Comment: This has to use the city_data_df data because the created merge data overcounts drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fares = total_fares/total_rides
average_fares
```

```python
# 5. Get the average fare per driver for each city type. 
average_fares_driver = total_fares/total_drivers
average_fares_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides": total_rides,
    "Total Drivers": total_drivers,
    "Total Fares": total_fares,
    "Average Fare per Ride": average_fares,
    "Average Fare per Driver": average_fares_driver})
pyber_summary_df
#David Comment: Straightforward, creates a df combining the elements that we made earlier
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
city_totalrides = pyber_data_df.groupby(["type","date"]).sum()["fare"]
city_totalrides
#David Comment: This time we use the groupby function on two variables, "type" and "date"
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
city_totalrides = city_totalrides.reset_index()
city_totalrides
#David Comment: recreates the index to allow pivot fuction to be used later
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
city_totalrides_pivot = city_totalrides.pivot(index = "date", columns = "type", values = "fare")
city_totalrides_pivot
#David Comment: acts similar to a pivot table in excel, makes the columns the city types and makes the rows the date with the data being the fare value
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
city_totalrides_weeks = city_totalrides_pivot.loc['2019-01-01':'2019-04-29']
city_totalrides_weeks
#David Comment: uses a loc function to narrow the range of dates
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
city_totalrides_weeks.index = pd.to_datetime(city_totalrides_weeks.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
city_totalrides_weeks.info()
#David Comment: top shows DatetimeIndex
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
city_totalrides_weeks = city_totalrides_weeks.resample("W").sum()
city_totalrides_weeks
#David Comment: use the resample function to combine the various date values into weeks
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

import datetime
import matplotlib.dates as mdates
#David Comment: we will need this to do formatting

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(18, 6))
ax.plot(city_totalrides_weeks)
ax.set_title('Total Fare by City Type')
ax.set_ylabel('Fare($USD)')
ax.legend(["Rural","Suburban","Urban"], loc = 'center', title = 'type')
#David Comment: normal object oriented plotting, creates a y label and a legend

#David Comment: I don't like you made me google how to format axes dates below
ax.set_xlim(datetime.date(2019, 1, 1), datetime.date(2019, 4, 29))
monthly = mdates.MonthLocator(interval=1)
ax.xaxis.set_major_locator(monthly)
labels = ['Jan\n2019','Feb','Mar','Apr']
ax.set_xticklabels(labels)
#David Comment: Tightens the interval shown on the graph, changes the ticks from weeks to months, changes tick labels to desired
#David Comment: Just so that it can match the desired graph as much as possible.

plt.savefig("analysis/PyBer_fare_summary.png")
```
